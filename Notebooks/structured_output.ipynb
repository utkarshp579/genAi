```python
text = "Hello, My Name is Utkarsh Mishra, "\
"Email is utkarshp5793@gmail.com"
```

### Structured Output


```python
from pydantic import BaseModel , Field
from typing import List

class ResponseStructure(BaseModel):
    name:str = Field(description="Complete Name")
    email:str = Field(description="Email Address")
    age:int = Field(description="Age")

structured_llm = llm.with_structured_output(ResponseStructure)
```

```python
res = structured_llm.invoke(f"Please give only name , email and age from the {text}")

# giving output as Class object of Response Structured created. 
```

```python
res   
```

```python
res.model_dump()   #  generate answer in dictionary form
```

```python
class Movie(BaseModel):
    title:str = Field(description="Movie Name")
    age:int = Field(description="Year")

class AllMovies(BaseModel):
    movies:List[Movie]

movie_llm = llm.with_structured_output(AllMovies)
```

```python
res = movie_llm.invoke("Give me three Sci - Fi movie")
```
