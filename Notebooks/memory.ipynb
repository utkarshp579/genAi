```python
from dotenv import load_dotenv
load_dotenv()

from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
```

```python
prompts = []        # list of dictionary

while True :
    query = input("User : ")
    if query.lower() in [ "stop" , "quit" , "end"]: 
        print(f"Bye")
        break

    prompts.append({"role" : "user" , "content" : query})

    res =   llm.invoke(prompts)
    prompts.append({"role" : "ai" , "content" : res.content})
    print(res.content)


# res.content
```
