#### Load Env Variables


```python
from dotenv import load_dotenv
import os 
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.output_parsers import StrOutputParser
```

```python
load_dotenv()
google_api_key = os.getenv("GOOGLE_API_KEY")
```

```python
os.environ["GOOGLE_API_KEY"] = google_api_key
llm_model = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
```

```python
# prompts = [
#     ("system" , "You are python developer."),
#     ("user" , "How to sort the array?")
# ]

prompts = [
    {"role" : "system" , "content" : "You are a Hindi translator." },
    {"role" : "user" , "content" : "I love to play Cricket. "}
]
```

```python
response = llm_model.invoke(prompts)
print(response.content)
```
